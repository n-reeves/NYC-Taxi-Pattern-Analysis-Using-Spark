## Task 8

Analysis of the relationship between time duration and distance for the yellow taxi service in January 2023

````python
# duration in hours along the x-axis and trip distance along the y-axis

plt.scatter(t8_yl['trip_duration_hours'], t8_yl['trip_distance'])
plt.xlabel('Duration (Hrs)')
plt.ylabel('Distance (Miles)')
plt.title('Yellow Taxi Trips by Day in January 2023')
plt.show()
````

````python
#skewed by outliers
#considering yonkers to south of staten island is about 50 miles, I will exclude outliers with distance over 100 miles
#Note that as this is not a part of the requirement for the assignement I will do this filtering in pandas
#I will also exclude trips longer than 10 hours as this is enough time to reach charlotte north carolina from NYC
#It is extremely unlikely that a journey could take this long unless the rider was going back and forth across the city continuously
fig, ax = plt.subplots()

t8_yl_filt = t8_yl[(t8_yl.trip_distance < 100)&(t8_yl.trip_duration_hours < 10)]
ax.scatter(t8_yl_filt['trip_duration_hours'], t8_yl_filt['trip_distance'])
ax.set_xlabel('Duration (Hrs)')
ax.set_ylabel('Distance (Miles)')
fig.suptitle('Yellow Taxi Trips by Day in January 2023')
fig.set_figwidth(10)
fig.set_figheight(5)
fig.show()
````
